```python
def decrypt(cipher, shift):
    result = ""
    for ch in cipher:
        if 'a' <= ch <= 'z':
            result += chr((ord(ch) - ord('a') - shift + 26) % 26 + ord('a'))
        elif 'A' <= ch <= 'Z':
            result += chr((ord(ch) - ord('A') - shift + 26) % 26 + ord('A'))
        else:
            result += ch
    return result


def main():
    cipher = "odroboewscdrolocdcwkbdmyxdbkmdzvkdpybwyeddrobo"
    print("Caesar Cipher Breaker\n")
    print(f"Original Cipher: {cipher}\n")
    print("Trying all possible shifts:")

    for shift in range(26):
        decrypted = decrypt(cipher, shift)
        print(f"Shift {shift}: {decrypted}")


if __name__ == "__main__":
    main()
```
